<h1>Spatial Averaging of Image Pixels Using 3×3, 10×10, and 20×20 Kernels</h1>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def average_blur_color_with_original(image_path, sizes=[3, 10, 20]):
    img = cv2.imread(image_path)  # Load color image
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB for matplotlib

    # Total plots = original + number of blur levels
    total_images = 1 + len(sizes)

    # Create 1-row subplot with total_images columns
    fig, axes = plt.subplots(1, total_images, figsize=(15, 5))
    fig.suptitle("Original and Average Blurred Images", fontsize=18)

    # Show original image
    axes[0].imshow(img)
    axes[0].set_title("Original")
    axes[0].axis('off')

    # Show blurred images in the rest of the subplots
    for i, k in enumerate(sizes):
        blurred = cv2.blur(img, (k, k))
        ax = axes[i + 1]  # Shift by 1 to skip original slot
        ax.imshow(blurred)
        ax.set_title(f"{k}x{k} Average")
        ax.axis('off')

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

average_blur_color_with_original(img_path)
```
