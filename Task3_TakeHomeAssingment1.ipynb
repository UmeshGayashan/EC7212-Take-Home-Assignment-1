<h2>Rotate an image by 45 and 90 degrees

```python
import cv2
import matplotlib.pyplot as plt

def rotate_image(image_path):
    img_read = cv2.imread(image_path)
    img = cv2.cvtColor(img_read, cv2.COLOR_BGR2RGB) # Convert BGR to RGB for correct color display in matplotlib
    if img is None:
        print(f"Failed to load image: {image_path}")
        return

    # 90 degrees
    rotated_90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

    # 45 degrees
    (h, w) = img.shape[:2] # First two values: height and width
    center = (w // 2, h // 2) # Center point of the image
    M = cv2.getRotationMatrix2D(center, 45, 1.0) # Affine transformation matrix
    rotated_45 = cv2.warpAffine(img, M, (w, h))

    # Show all images side by side
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(img)
    axes[0].set_title('Original')
    axes[0].axis('off')

    axes[1].imshow(rotated_90)
    axes[1].set_title('Rotated 90')
    axes[1].axis('off')

    axes[2].imshow(rotated_45)
    axes[2].set_title('Rotated 45')
    axes[2].axis('off')

    plt.tight_layout() # Automatically adjusts the spacing between plot elements
    plt.show()

rotate_image(img_path)
```
